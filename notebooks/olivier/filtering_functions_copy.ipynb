### Function

```python
recipe_df = recipe_df['tags'].map(lambda x: x if 'cod' in x)
```

```python
MEAT_LIST = ['meat','bacon', 'sausage', 'pork', 'chicken', 'beef', 'duck', 'lamb', 'animal', 'chorizo',\
            'bushmeat', 'horse', 'goose', 'meatball', 'lean', 'poultry', 'turkey', 'veal', 'steak', 'pig',\
            'red meat', 'fillet', 'ham', 'mutton', 'rib', 'cutlet', 'sheep', 'biltong']
```

```python
#LIST_DIET = ['classic', 'low carb', 'flexitarian', 'vegetarian', 'pescetarian', 'vegan']

def diet_filter(diet, df):
    if diet == 'classic':
        return df
    elif diet == 'vegan':
        return df[df["tags"].str.contains(diet)]
    elif diet == 'vegetarian':
        return df[df["tags"].str.contains(diet)]
    elif diet == 'pescetarian':
        for meat in MEAT_LIST:
            df = df[(~df["tags"].str.contains(meat))&(~df["ingredients"].str.contains(meat))]
        return df
    elif diet == 'low-carb':
        return df[df["tags"].str.contains(diet)]
```

```python
def allergie_filter(allergies, allergies_input, df):
    
    for key, value in allergies.items():
        if key not in allergies_input:
            allergies[key] = []        

    #allergies
    for key, value in allergies.items():
        for i in value:
            df = df[~df["ingredients"].str.contains(i)]
    return df
```

## Time

- Up to 15min
- Up to 30min
- Up to 45min
- Up to 1h
- More than 1h

### Function

```python
def time_filter(time, df):
    if time == 'up to 15 min':
        return df[df.minutes<16]
    elif time == 'up to 30 min':
        return df[df.minutes<31]
    elif time == 'up to 45 min':
        return df[df.minutes<46]
    elif time == 'up to 1h':
        return df[df.minutes<61]
    elif time == 'more than 1h':
        return df
```
