# Prefs

```python
cols = ['app_user_id', 'name', 'goal', 'diet', 'allergies', 'dislikes', 'custom_dsl']
user_prefs = pd.DataFrame(columns=cols)
```

```python
def add_user(info):
    global user_prefs
    
    new_user = {'app_user_id': info[0], 
                'name': info[1], 
                'goal': info[2], 
                'diet': info[3], 
                'allergies': info[4],
                'dislikes': info[5],
                'custom_dsl': info[6]}
    
    user_prefs = user_prefs.append(new_user, ignore_index=True)
```

```python
jess = [1, 'Jessica Ribeiro', 'lose', 'classic', 'tree_nuts', 'cauliflower, brussels sprouts, tofu', 'blood sausage, nutmeg, spinach']
nuno = [2, 'Nuno Loureiro', 'lose', 'low-carb', 'shellfish', 'beets, brussels sprouts, mushrooms, shrimp, tofu', 'lures, raisin, bean, grain, avocado, broccoli']
phillip = [3, 'Phillip Zastrow', 'build', 'vegetarian', 'mustard', 'turnips, quinoa, beets', 'raisin, pineapple']
olivier = [4, 'Olivier Paulo', 'maintain', 'vegan', 'None', 'brussels sprouts, blue cheese', 'sobriety, artichoke']
```

```python
users = [jess, nuno, phillip, olivier]

for user in users:
    add_user(user)
```
