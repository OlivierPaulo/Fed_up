The first target is to obtain a ordered list of all the possible ingredients retreived from the file.

```python
ingredients = data["ingredients"]
```

# The ingredients consist of a string that looks like a list.
Approach below:
- split the string by ","
- replace fragements like "[", "]", "'" with "" (>29500 items)
- remove duplicates after cleaning the strings -> now 14902 items

```python
ingredients_list = []
for i in ingredients:
    for j in i.split(","):
        if j not in ingredients_list:
            ingredients_list.append(j)
```

```python
ingredients_list = [i.replace("[","").replace("]","").replace("'","").replace('"','').strip() for i in ingredients_list]
```

```python
clean_ingredients_list = []
for i in ingredients_list:
    if i not in clean_ingredients_list:
        clean_ingredients_list.append(i)
```

## changing the ingredient into a list, not a huge string for searchability

```python
ingredients_temp = data["ingredients"]
ingred_temper = []
```

```python
for ing in ingredients_temp:
    list_temp = []
    for j in ing.split(","):
        list_temp.append(j.replace("[","").replace("]","").replace("'","").replace('"','').strip())
    ingred_temper.append(list_temp)
```

```python
data["ingredients"]=ingred_temper
```
