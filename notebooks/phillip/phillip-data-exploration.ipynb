# The ingredients consist of a string that looks like a list.
Approach below:
- split the string by ","
- replace fragements like "[", "]", "'" with "" (>29500 items)
- remove duplicates after cleaning the strings -> now 14902 items

```python
ingredients_list = []
for i in ingredients:
    for j in i.split(","):
        if j not in ingredients_list:
            ingredients_list.append(j)
```

```python
ingredients_list = [i.replace("[","").replace("]","").replace("'","").replace('"','').strip() for i in ingredients_list]
```

```python
clean_ingredients_list = []
for i in ingredients_list:
    if i not in clean_ingredients_list:
        clean_ingredients_list.append(i)
```

```python
tags = data["tags"]
```

```python
tag_list_clean = []
tag_list = []
for i in tags:
    for j in i.split(","):
        ji = j.replace("'", "").replace(" ","").replace("]", "").replace("[", "")
        tag_list.append(ji)
        if ji not in tag_list_clean:
            tag_list_clean.append(ji)
```

```python
tag_list_clean
```

```python
len(tag_list), len(tag_list_clean)
```

```python
def count_occurrence(tag_clean,tag_complete, threshold):
    exclude =[]
    for i in tag_clean:
        count = 0
        for j in tag_complete:
            if j == i:
                count+=1
        if count < threshold:
            exclude.append(i)
    return exclude
    
```

```python
exclude = count_occurrence(tag_list_clean, tag_list, 10)
exclude
```

```python
len(exclude)
```
